```python
import os
import numpy as np
import pydicom
import cv2
import torch
import torch.nn as nn
import torch.optim as optim
import torch.utils.data as data
import torch.nn.functional as F
from torchinfo import summary
import random
import h5py
import torchvision.transforms as transforms
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim
import matplotlib.pyplot as plt
```

```python
# Configurar dispositivo
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Usando dispositivo: {device}")
```

```python
dicom_root_dir = './Imagenes_Finales/176X256X256'
PATCH_SIZE = 64  # Tamaño de los patches 3D
# Ruta donde se guardarán los modelos y métricas
checkpoint_dir = "./checkpoints"
os.makedirs(checkpoint_dir, exist_ok=True)
```

```python
# Función para cargar y normalizar un volumen 3D
def load_dicom_volume(dicom_dir):
    dicom_files = sorted([os.path.join(dicom_dir, f) for f in os.listdir(dicom_dir) if f.endswith(".dcm")],
                         key=lambda f: float(pydicom.dcmread(f).InstanceNumber))
    volume = np.array([pydicom.dcmread(f).pixel_array for f in dicom_files], dtype=np.float32)
    volume = (volume - np.min(volume)) / (np.max(volume) - np.min(volume) + 1e-7)  # Normalización 0-1
    return volume
```

```python
# Función para extraer un parche aleatorio
def extract_random_patch(volume, patch_size=PATCH_SIZE):
    d, h, w = volume.shape
    d_start = random.randint(0, d - patch_size)
    h_start = random.randint(0, h - patch_size)
    w_start = random.randint(0, w - patch_size)
    patch = volume[d_start:d_start+patch_size, h_start:h_start+patch_size, w_start:w_start+patch_size]
    return patch
```

```python
# Generar parche en baja resolución
def low_res_patch(patch, factor=0.5):
    d, h, w = patch.shape
    low_res = np.array([cv2.resize(cv2.resize(slice, 
                                              (int(w * factor), int(h * factor)), interpolation=cv2.INTER_AREA),
                                   (w, h), interpolation=cv2.INTER_LINEAR) 
                        for slice in patch])
    return low_res
```

```python
# Dividir estudios en 90% entrenamiento y 10% validación
study_dirs = [os.path.join(dicom_root_dir, d) for d in os.listdir(dicom_root_dir) if os.path.isdir(os.path.join(dicom_root_dir, d))]
random.seed(42)
random.shuffle(study_dirs)
split_idx = int(0.9 * len(study_dirs))
train_dirs = study_dirs[:split_idx]
val_dirs = study_dirs[split_idx:]
```

```python

class MRIDataset(data.Dataset):
    def __init__(self, study_dirs, num_patches=20, scale_factor=0.5):
        self.study_dirs = study_dirs
        self.num_patches = num_patches
        self.scale_factor = scale_factor

    def __len__(self):
        return len(self.study_dirs) * self.num_patches

    def __getitem__(self, idx):
        study_idx = idx // self.num_patches
        volume = load_dicom_volume(self.study_dirs[study_idx])
        hr_patch = extract_random_patch(volume)
        lr_patch = low_res_patch(hr_patch, factor=self.scale_factor)
        lr_patch = torch.tensor(lr_patch, dtype=torch.float32).unsqueeze(0)
        hr_patch = torch.tensor(hr_patch, dtype=torch.float32).unsqueeze(0)
        return lr_patch, hr_patch
```

```python
# Loaders
train_dataset = MRIDataset(train_dirs)
val_dataset = MRIDataset(val_dirs)
train_loader = data.DataLoader(train_dataset, batch_size=4, shuffle=True, num_workers=0)
val_loader = data.DataLoader(val_dataset, batch_size=4, shuffle=False, num_workers=0)  
```

```python
# Modelo cGAN 3D - Generador
class ResidualBlock(nn.Module):
    def __init__(self, in_channels):
        super(ResidualBlock, self).__init__()
        self.conv1 = nn.Conv3d(in_channels, in_channels, kernel_size=3, padding=1)
        self.bn1 = nn.BatchNorm3d(in_channels)
        self.conv2 = nn.Conv3d(in_channels, in_channels, kernel_size=3, padding=1)

    def forward(self, x):
        residual = x
        x = F.leaky_relu(self.bn1(self.conv1(x)), 0.2)
        x = self.conv2(x)  
        return x + residual

class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()
        self.conv1 = nn.Conv3d(1, 32, kernel_size=3, padding=1) 
        self.res_blocks = nn.Sequential(*[ResidualBlock(32) for _ in range(3)])  
        self.conv2 = nn.Conv3d(32, 1, kernel_size=3, padding=1)
        
    def forward(self, x):
        x = F.leaky_relu(self.conv1(x), 0.2)
        x = self.res_blocks(x)
        x = torch.sigmoid(self.conv2(x))
        return x
```

```python
# Modelo cGAN 3D - Discriminador
class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        self.conv1 = nn.Conv3d(2, 32, kernel_size=3, padding=1)  
        self.conv2 = nn.Conv3d(32, 64, kernel_size=3, padding=1)  
        self.conv3 = nn.Conv3d(64, 128, kernel_size=3, padding=1)  
        self.pool = nn.AdaptiveAvgPool3d(1)
        self.fc = nn.Linear(128, 1) 

    def forward(self, lr, hr):
        x = torch.cat([lr, hr], dim=1)
        x = F.leaky_relu(self.conv1(x), 0.2)
        x = F.leaky_relu(self.conv2(x), 0.2)
        x = F.leaky_relu(self.conv3(x), 0.2)
        x = self.pool(x).view(x.size(0), -1)
        x = torch.sigmoid(self.fc(x))
        return x
```

```python
# Inicializar modelos
generator = Generator().to(device)
discriminator = Discriminator().to(device)
```

```python
# Summary del Generador
summary(generator, input_size=(1, 1, 64, 64, 64)) 
```

```python
# Summary del Discriminador
summary(discriminator, input_size=[(1, 1, 64, 64, 64), (1, 1, 64, 64, 64)])  
```

```python
# Definir función de pérdida y optimizadores
criterion = nn.MSELoss()
criterion_mse = nn.MSELoss()
optimizer_G = optim.AdamW(generator.parameters(), lr=1e-4, betas=(0.0, 0.9))
optimizer_D = optim.AdamW(discriminator.parameters(), lr=1e-4, betas=(0.0, 0.9))
```

```python
# Archivo para guardar métricas

h5_file = os.path.join(checkpoint_dir, "metrics.h5")
with h5py.File(h5_file, "w") as f:
    for name in ["loss_G", "loss_D", "psnr", "ssim", "val_psnr", "val_ssim"]:  
        f.create_dataset(name, (0,), maxshape=(None,))
```

```python
num_epochs = 50

print("Entrenamiento Iniciado.")
for epoch in range(num_epochs):
    generator.train()
    total_psnr, total_ssim, count = 0, 0, 0
    for lr_patch, hr_patch in train_loader:
        lr_patch, hr_patch = lr_patch.to(device), hr_patch.to(device)
        batch_size = lr_patch.size(0)

        # Generador
        optimizer_G.zero_grad()
        fake_patch = generator(lr_patch)
        pred_fake = discriminator(lr_patch, fake_patch)
        loss_G = torch.mean((pred_fake - 1) ** 2) + criterion_mse(fake_patch, hr_patch) * 3
        loss_G.backward()
        optimizer_G.step()

        # Discriminador
        optimizer_D.zero_grad()
        pred_real = discriminator(lr_patch, hr_patch)
        pred_fake = discriminator(lr_patch, fake_patch.detach())
        loss_D = 0.5 * (torch.mean((pred_real - 1) ** 2) + torch.mean(pred_fake ** 2))
        loss_D.backward()
        optimizer_D.step()

        # Métricas entrenamiento
        for i in range(batch_size):
            fake_np = fake_patch[i].cpu().detach().numpy().squeeze()
            hr_np = hr_patch[i].cpu().detach().numpy().squeeze()
            total_psnr += psnr(hr_np, fake_np, data_range=1)
            total_ssim += ssim(hr_np, fake_np, data_range=1)
            count += 1

    avg_psnr = total_psnr / count
    avg_ssim = total_ssim / count

    # === VALIDACIÓN ===
    generator.eval()
    with torch.no_grad():
        val_total_psnr, val_total_ssim, val_count = 0, 0, 0
        for lr_patch, hr_patch in val_loader:
            lr_patch, hr_patch = lr_patch.to(device), hr_patch.to(device)
            fake_patch = generator(lr_patch)
            fake_np = fake_patch[0].cpu().numpy().squeeze()
            hr_np = hr_patch[0].cpu().numpy().squeeze()
            val_total_psnr += psnr(hr_np, fake_np, data_range=1)
            val_total_ssim += ssim(hr_np, fake_np, data_range=1)
            val_count += 1
        avg_val_psnr = val_total_psnr / val_count
        avg_val_ssim = val_total_ssim / val_count
    # ==================

    print(f"Epoch [{epoch+1}/{num_epochs}] - Loss G: {loss_G.item():.4f}, Loss D: {loss_D.item():.4f}, "
          f"PSNR: {avg_psnr:.2f}, SSIM: {avg_ssim:.4f}, VAL_PSNR: {avg_val_psnr:.2f}, VAL_SSIM: {avg_val_ssim:.4f}")

    with h5py.File(h5_file, "a") as f:
        for key, value in zip(["loss_G", "loss_D", "psnr", "ssim", "val_psnr", "val_ssim"],
                              [loss_G.item(), loss_D.item(), avg_psnr, avg_ssim, avg_val_psnr, avg_val_ssim]):
            dataset = f[key]
            dataset.resize((dataset.shape[0] + 1,))
            dataset[-1] = value

    torch.save(generator.state_dict(), os.path.join(checkpoint_dir, f"generator_epoch_{epoch+1}.pth"))
    torch.save(discriminator.state_dict(), os.path.join(checkpoint_dir, f"discriminator_epoch_{epoch+1}.pth"))

print("Entrenamiento completado.")
```
